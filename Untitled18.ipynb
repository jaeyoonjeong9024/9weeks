```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# 1. 데이터 로드 및 전처리
# CSV에 헤더가 없으므로 헤더 이름을 지정합니다.
column_names = ['user_id', 'activity', 'timestamp', 'x_accel', 'y_accel', 'z_accel']
df = pd.read_csv('/content/WISDM.csv', header=None, names=column_names)

# 'Jogging'과 'Sitting' 활동만 필터링
df = df[(df['activity'] == 'Jogging') | (df['activity'] == 'Sitting')].copy()

# 'Class' 열 생성 (Sitting=0: 정상, Jogging=1: 이상)
df['Class'] = df['activity'].apply(lambda x: 0 if x == 'Jogging' else 1)
# 원본 코드의 'Time' 열처럼 불필요한 열 제거
df = df.drop(['user_id', 'activity', 'timestamp'], axis=1)

# 결측치 확인
print(f"결측치 존재 여부: {df.isnull().values.any()}")
print(f"데이터 형태: {df.shape}")
```

```python
#
LABELS = ["Jogging", "Sitting"]
count_classes = pd.value_counts(df['Class'], sort = True)

count_classes.plot(kind = 'bar', rot=0)
plt.title("Activity class distribution")
plt.xticks(range(2), LABELS)
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```

```python
# 원본 코드의 'Amount' 스케일링과 동일하게 가속도계 데이터를 스케일링합니다.
scaler = StandardScaler()
accel_cols = ['x_accel', 'y_accel', 'z_accel']
df[accel_cols] = scaler.fit_transform(df[accel_cols])

# 원본 코드와 동일한 로직으로 데이터 분할
# (전체 데이터를 먼저 분할)
RANDOM_SEED = 42
X_train_df, X_test_df = train_test_split(df, test_size=0.2, random_state=RANDOM_SEED)

# 훈련 데이터는 '정상'(Sitting, Class=0) 데이터만 사용
X_train = X_train_df[X_train_df.Class == 0]
X_train = X_train.drop(['Class'], axis=1)

# 테스트 데이터 준비
y_test = X_test_df['Class']
X_test = X_test_df.drop(['Class'], axis=1)

# .values로 numpy 배열 변환
X_train = X_train.values
X_test = X_test.values

print(f"X_train (정상) 형태: {X_train.shape}")
print(f"X_test (전체) 형태: {X_test.shape}")
```

```python
# 2. 오토인코더 모델
input_dim = X_train.shape[1] # 3 (x, y, z)
encoding_dim = 2  # 원본 14 대신 3개 피처에 맞게 축소 (3 -> 2 -> 1)
hidden_dim = int(encoding_dim / 2) # 1

input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu')(input_layer)
encoded = Dense(hidden_dim, activation='relu')(encoded) # 1차원
decoded = Dense(hidden_dim, activation='relu')(encoded) # 1차원
decoded = Dense(input_dim, activation='sigmoid')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

autoencoder.summary()
```

```python
# 3. 모델 학습
history = autoencoder.fit(X_train, X_train,
                          epochs=50,
                          batch_size=256,
                          shuffle=True,
                          validation_split = 0.1,
                          verbose=1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

```python
# 4. 평가
predictions = autoencoder.predict(X_test)
mse = np.mean(np.power(X_test - predictions, 2), axis=1)

error_df = pd.DataFrame({'reconstruction_error': mse,
                         'true_class': y_test})

print(error_df.describe())
error_df.head()
```

```python
# threshold 값을 조정하여 최선의 정확도를 얻음, 그래프로를 보고 결정
# (WISDM 데이터의 오류 분포를 보고 0.3으로 설정)
threshold = 0.3

groups = error_df.groupby('true_class')
fig, ax = plt.subplots(figsize=(12, 6))

for name, group in groups:
    ax.plot(group.index, group.reconstruction_error, marker='o', ms=3.5, linestyle='',
            label= "Sitting" if name == 1 else "Jogging")

ax.hlines(threshold, ax.get_xlim()[0], ax.get_xlim()[1], colors="r", zorder=100, label='Threshold')
ax.hlines(threshold, ax.get_xlim()[0], ax.get_xlim()[1], colors="r", zorder=100, label='Threshold')
ax.legend()
plt.title("Reconstruction error for different classes")
plt.ylabel("Reconstruction error")
plt.xlabel("Data point index")
plt.show()
```

```python
# 5. 혼동 행렬 및 분류 리포트
y_pred = [1 if e > threshold else 0 for e in error_df.reconstruction_error.values]
conf_matrix = confusion_matrix(error_df.true_class, y_pred)

plt.figure(figsize=(8, 8))
sns.heatmap(conf_matrix, xticklabels=LABELS, yticklabels=LABELS, annot=True, fmt="d", cmap='Blues');
plt.title("Confusion matrix")
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()

print("\n[ Classification Report ]")
print(classification_report(error_df.true_class, y_pred, target_names=LABELS))
```
